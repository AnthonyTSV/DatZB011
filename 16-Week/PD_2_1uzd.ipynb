\end{align}

Protams, ka zāle nav bezgalīgs resurss, tādējādi, ja zaķu populācija kļūst pārlieku liela, tā var sastapties ar zāles trūkumu. Varam ieviest šo ideju modelī, pieņemot, ka zāles trūkumu dēļ zaķi sāks mirt ar ātrumu $-aR^2/b$, kur $b>0$. Rezultātā, populācijas izmērus var aprakstīt ar vienādojumu:

\begin{align}
	\dot R=aR\left(1-\frac{R}{b}\right).
\end{align}

3. Kā izmainās zaķu populācija $R(t)$ laikā, ja ierobežo pieeju ēdienam?

Sākumā nepieciešams sadalīt mainīgos un nedaudz vienkāršot izteiksi, kā arī saucēju.
\begin{align}
    \dot{R}&=aR\left(1-\frac{R}b\right)
    & \frac{\mathrm{d}R}{R^2-bR}&=-\frac{a}{b}\mathrm{d}t
    & \frac{\mathrm{d}R}{R(R-b)}&=-\frac{a}{b}\mathrm{d}t\\
\end{align}

Sadalīsim daļskaitli elementārdaļās (*partial fraction decomposition*), lai kreiso pusu padarītu integrējamu.
$$
    \frac{\mathrm{d}R}{R(R-b)}
    =\frac{A\,\mathrm{d}R}R+\frac{B\,\mathrm{d}R}{R-b}
    =\begin{bmatrix}A+B=0 & B=\frac1b \\ -bA=1 & A=-\frac1b\end{bmatrix}
    =\frac1b\left(\frac{\mathrm{d}R}{R-b}-\frac{\mathrm{d}R}R\right)
$$

Visbeidzot iespējams integrēt un aizvietot sākotnējo nosacījumu uz $R(0)=R_0$.
\begin{align}
    \ln\frac{R-b}R&=-at+\ln C & R(t)&=\frac{b}{1-C\exp(-at)} & R(t)&=\frac{b}{1-\left(1-\frac{b}{R_0}\right)\exp(-at)}\\
    1-\frac{b}{R}&=C\exp(-at) & R_0&=\frac{b}{1-C}
\end{align}

Apskatīsim citus reālistiskus modeļus, atcerēsimies, ka zaķi nav vienīgie dzīvnieki ekosistēmā: ieviesīsim lapsas, kas ieradušās zaķus medīt. Turpināsim apzīmēt zaķu populāciju ar $R$ un apzīmēsim lapsu populāciju ar $F$. Šajā uzdevuma daļā pieņemsim, ak trušu populācija ir ierobežota, tāpēc zāles vienmēr pietiek.

Zaķu populacija turpina augt proporcionāli tās izmēram $R$, tomēr medību dēļ tā arī samazinās atbilstoši iespējai, ka lapsa nomedīs zaķi, resp., $\dot R=aR-cRF$.

Saistībā ar lapsām - jo biežāk iespējams atrast zaķi, ko apēst, jo populācija aug straujāk, bet bez ēdiena populācija samazinās atbilstoši tās izmēram $F$, resp., $\dot F=hRF-kF$, kur visiem parametriem $a$, $c$, $h$ un $k$ ir pozitīvas vērtības.
Sakotnējās zaķu un lapsu populācijas ir $R_0$ un $F_0$. 

6. Izslēdziet laiku un analītiski atrodiet $F(R)$ attiecību.

Lai tiktu vaļā no atkarības no laika, nepieciešams izdalīt abus vienādojumus.
\begin{align}
    \begin{cases}
        \frac{\mathrm{d}R}{\mathrm{d}t}=aR-cRF\\
        \frac{\mathrm{d}F}{\mathrm{d}t}=hRF-kF
    \end{cases}
    && \frac{\mathrm{d}F}{\mathrm{d}R}=\frac{hRF-kF}{aR-cRF}
\end{align}

Šo vienādojumu iespējams sadalīt, ja skaitītāju un saucēju izdala ar $RF$.
\begin{align}
    aR\,\mathrm{d}F-cRF\,\mathrm{d}F-hRF\,\mathrm{d}R+kF\,\mathrm{d}R&=0
    & \left(\frac{a}{F}-c\right)\mathrm{d}F&=\left(h-\frac{k}{R}\right)\mathrm{d}R\\
    a\frac{\mathrm{d}F}{F}-c\,\mathrm{d}F-h\,\mathrm{d}R+k\frac{\mathrm{d}R}{R}&=0
    & a\ln F-cF&=hR-k\ln R+C
\end{align}

Visbeidzot izmantojam, ka $R(t=0)=R_0$ un $F(t=0)=F_0$, lai noteiktu $C$. Šai izteiksmei nav analītiskā risinājumu, tāpēc atstāsim netiešo formu.
\begin{align}
    \begin{cases}
        a\ln F+k\ln R-cF-hR=C\\
        a\ln F_0+k\ln R_0-cF_0-hR_0=C\\
    \end{cases}
    && a\ln\frac{F}{F_0}+k\ln\frac{R}{R_0}&=c(F-F_0)+h(R-R_0)
\end{align}

- _0 scenārijs_. Vienā sākotnējā populāciju izmērā ekosistēma ir līdzsvarā. Lapsas apēd zaķus tieši tajā pašā ātrumā, kā zaķi vairojas. Tas notiek, kad $\dot{R}=0$ un $\dot{F}=0$, t.i., kad $F=a/c$ un $R=k/h$.
