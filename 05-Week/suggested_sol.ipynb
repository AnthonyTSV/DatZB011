# Uzdevumu atrisinājumu piemēri

## 1. uzdevums

```python
import matplotlib.pyplot as plt
import numpy as np

v0 = 0
a = 2
t = np.arange(0, 10.1, 0.1)

v = v0 + a * t
l = v0 * t + 0.5 * a * t**2

plt.figure(figsize=(10,6))
plt.plot(t, v, label='Ātrums (m/s)', color='blue')
plt.plot(t, l, label='Ceļš (m)', color='red')

plt.title('Ķermeņa kustības ātrums un ceļš')
plt.xlabel('Laiks (s)')
plt.ylabel('Ātrums / Ceļš')
plt.legend()
plt.grid(True)
plt.savefig('1uzd.png', dpi=300)
plt.show()
```

## 3. uzdevums

```python
import matplotlib.pyplot as plt
import numpy as np

v0 = 50
theta_deg = 30
theta = np.radians(theta_deg)
g = 9.81
t = np.arange(0, 10.1, 0.1)

v0x = v0 * np.cos(theta)
v0y = v0 * np.sin(theta)

x = v0x * t
y = v0y * t - 0.5 * g * t**2

vx = v0x * np.ones_like(t)
vy = v0y - g * t

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=False)

ax1.plot(x, y, color='green', linestyle='-', marker='o')
ax1.set_title('Slīpi pret horizontu sviesta ķermeņa kustības trajektorija')
ax1.set_xlabel('x (m)')
ax1.set_ylabel('y (m)')
ax1.grid(True, linestyle='--', alpha=0.7)

ax2.plot(t, vx, label='Ātruma projekcija pa x asi (m/s)', color='blue', linestyle='-')
ax2.plot(t, vy, label='Ātruma projekcija pa y asi (m/s)', color='red', linestyle='--')
ax2.set_title('Ātruma projekciju izmaiņa laikā')
ax2.set_xlabel('Laiks (s)')
ax2.set_ylabel('Ātrums (m/s)')
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.7)

fig.suptitle('Slīpi pret horizontu sviesta ķermeņa kustības analīze', fontsize=16)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('3uzd.png', dpi=300)
plt.show()
```

## 5. uzdevums

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

n = 1
R = 0.0821
V = np.arange(1, 10.5, 0.5)
temperatures = [300, 310, 320, 330, 340]

V_list = []
P_list = []
T_list = []

for T in temperatures:
    P = (n * R * T) / V
    V_list.extend(V)
    P_list.extend(P)
    T_list.extend([T] * len(V))

T_array = np.array(T_list)

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

scatter = plt.scatter(
    V_list, P_list,
    c=T_list,
    cmap='viridis',
    s=100,
    alpha=0.7,
    edgecolors='w',
)

cbar = plt.colorbar(scatter)
cbar.set_label('Temperatūra (K)', fontsize=12)

plt.title('Ideālā gāzes spiediena un tilpuma attiecība dažādos temperatūras punktos', fontsize=16)
plt.xlabel('Tilpums (L)', fontsize=14)
plt.ylabel('Spiediens (atm)', fontsize=14)
plt.tight_layout()
plt.savefig('5_uzd.png', dpi=300)
plt.show()
```
